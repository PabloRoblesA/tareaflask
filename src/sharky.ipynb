```python
# Visualizacion del dataset
import pandas as pd

df = pd.read_csv('../data/raw/Sharks_California.csv')
df
```

```python
df.info()
df.describe()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
#EDA
import matplotlib.pyplot as plt
import seaborn as sns

#Graficar variables categoricas
fig,axis = plt.subplots(3, 3, figsize=(10,8))

sns.histplot(ax=axis[0,0],data=df,x='County')             
sns.histplot(ax=axis[0,1],data=df,x='Location').set_xticks([])
sns.histplot(ax=axis[0,2],data=df,x='Mode')
sns.histplot(ax=axis[1,0],data=df,x='Injury')
sns.histplot(ax=axis[1,1],data=df,x='InjuryT')
sns.histplot(ax=axis[1,2],data=df,x='Depth')
sns.histplot(ax=axis[2,0],data=df,x='Species')

plt.tight_layout()
plt.show()
```

```python
#Analisis multivariante: Categorico-Categorico
sns.countplot(data=df,x='InjuryT',hue='Mode')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df,x='InjuryT',hue='Species')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df,x='InjuryT',hue='County')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
df.columns
```

```python
df=df[['County', 'Location', 'Mode', 'Injury','InjuryT', 'Depth', 'Species', 'Longitude', 'Latitude']]
```

```python
#Pasar variables categoricas a numericas
df['County_n']=pd.factorize(df['County'])[0]                    
df['Location_n']=pd.factorize(df['Location'])[0]
df['Mode_n']=pd.factorize(df['Mode'])[0]
df['Injury_n']=pd.factorize(df['Injury'])[0]
df['InjuryT_n']=pd.factorize(df['InjuryT'])[0]
df['Depth_n']=pd.factorize(df['Depth'])[0]
df['Species_n']=pd.factorize(df['Species'])[0]

df.head()
```

```python
df.columns
```

```python
# Analisis de correlacion.
sns.heatmap(df[['Longitude', 'Latitude', 'County_n', 'Location_n', 'Mode_n', 'Injury_n','InjuryT_n', 'Depth_n', 'Species_n']].corr(),annot=True,fmt='.2f')

plt.tight_layout()
plt.show()
```

```python
# Revision de distribucion y outlayers.
fig,axis = plt.subplots(3, 3, figsize=(10,8))

sns.boxplot(ax=axis[0,0],data=df,y='Longitude')
sns.boxplot(ax=axis[0,1],data=df,y='Latitude')       
sns.boxplot(ax=axis[0,2],data=df,y='County_n')       
sns.boxplot(ax=axis[1,0],data=df,y='Location_n')
sns.boxplot(ax=axis[1,1],data=df,y='Mode_n')
sns.boxplot(ax=axis[1,2],data=df,y='Injury_n')
sns.boxplot(ax=axis[2,0],data=df,y='InjuryT_n')
sns.boxplot(ax=axis[2,1],data=df,y='Depth_n')
sns.boxplot(ax=axis[2,2],data=df,y='Species_n') 

plt.tight_layout()
plt.show()
```
