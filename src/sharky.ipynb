```python
# Visualizacion del dataset
import pandas as pd

df = pd.read_csv('../data/raw/Sharks_California.csv')
df
```

```python
df.info()
df.describe()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df
```

```python
#EDA
import matplotlib.pyplot as plt
import seaborn as sns

#Graficar variables categoricas
fig,axis = plt.subplots(3, 3, figsize=(10,8))

sns.histplot(ax=axis[0,0],data=df,x='County')             
sns.histplot(ax=axis[0,1],data=df,x='Location').set_xticks([])
sns.histplot(ax=axis[0,2],data=df,x='Mode')
sns.histplot(ax=axis[1,0],data=df,x='Injury')
sns.histplot(ax=axis[1,1],data=df,x='InjuryT')
sns.histplot(ax=axis[1,2],data=df,x='Depth')
sns.histplot(ax=axis[2,0],data=df,x='Species')

plt.tight_layout()
plt.show()
```

```python
#Analisis multivariante: Categorico-Categorico
sns.countplot(data=df,x='InjuryT',hue='Mode')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df,x='InjuryT',hue='Species')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df,x='InjuryT',hue='County')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
